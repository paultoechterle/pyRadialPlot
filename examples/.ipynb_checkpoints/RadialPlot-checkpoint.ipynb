```python
from pyRadialPlot import radialplot
import csv

with open("ressources/FTdata.csv", "r") as f:
    file = csv.reader(f)
    name = next(file)
    line = next(file)
    zeta, zeta_err = float(line[0]), float(line[1])
    line = next(file)
    rhod, rhod_err = float(line[0]), int(float(line[1]))
    
    Ns = []
    Ni = []
    dpars = []
    
    for line in file:
        Ns.append(int(line[0]))
        Ni.append(int(line[1]))
        dpars.append(float(line[2]))
```

```python
import matplotlib.pyplot as plt

fig = plt.Figure(figsize=(6,6))
radialplot(Ns, Ni, zeta, rhod)
```
